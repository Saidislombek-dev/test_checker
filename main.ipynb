```python
from checker import Check_answers
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# test keys

keys = ['A','B','C','D','E','F','A','B']
```

```python
# instruction

plt.figure(figsize=(3,3))
ins = cv2.cvtColor(cv2.imread("images/instruction.jpg"),cv2.COLOR_BGR2RGB)
plt.imshow(ins)
plt.show()
```

```python
# 1-test

plt.figure(figsize=(11,13))

img = cv2.cvtColor(cv2.imread("images/test1.jpg"),cv2.COLOR_BGR2RGB)

checked_img, cor_ans = Check_answers(img, keys)

plt.subplot(121)
plt.title("Real image")
plt.imshow(img)
plt.subplot(122)
plt.title("Checked image")
plt.imshow(checked_img)
plt.show()
print("Result:",round(cor_ans/8*100),"%")
```
